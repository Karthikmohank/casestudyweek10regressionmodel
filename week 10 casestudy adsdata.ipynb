### Creating a regression model to predict profit

```python
y=data['Profit(in Thousands)']
X=data.drop(['Profit(in Thousands)','Date'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model1=lr.fit(X_train,y_train)
Prediction=model1.predict(X_test)
```

```python
print(Prediction)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is:',mean_squared_error(y_test,Prediction))
```

```python
## The multilinear regression model has an error of  4.64028148505357e-25
## which means the average sqared difference between observed and predicted values is  4.64028148505357e-25.
```

```python
r_sq=model1.score(X,y)
print("coefficient of determination is:",r_sq)
```

```python
print('y intercept:',model1.intercept_)
```

```python
print('coefficients of best fit  regression line is:',model1.coef_)
```

###  Predict the profit generated by the company

```python
pred_prof=model1.predict([[290,312,309,298,320]])
print('The predicted profit with given ads spending amount is:',pred_prof)
```
